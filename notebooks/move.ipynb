```python
import os
import shutil

print("Starting file copy...\n")

# SOURCE folder (where .pkl files exist)
source_folder = r"C:\Users\RAGHU\OneDrive - Vasavi College Of Engineering\Desktop\GITHUB\PROJECT-1\fraud-detection-system\data\processed"

# DESTINATION folder (your models folder)
destination_folder = r"C:\Users\RAGHU\OneDrive - Vasavi College Of Engineering\Desktop\GITHUB\PROJECT-1\fraud-detection-system\src\models"

# Files to copy
files_to_copy = [
    "final_model.pkl",
    "best_threshold.pkl",
    "feature_names.pkl",
    "shap_explainer.pkl"
]

# Copy files
for file in files_to_copy:

    source_path = os.path.join(source_folder, file)
    destination_path = os.path.join(destination_folder, file)

    if os.path.exists(source_path):

        shutil.copy(source_path, destination_path)
        print(f"✅ Copied: {file}")

    else:

        print(f"❌ File not found: {file}")

print("\n✅ All files processed!")
```
